### Scatterplot using Seaborn

```python
sns.relplot(x="ApplicantIncome", y="LoanAmount", data = df)
```

### SNS.relplot using Seaborn

```python
sns.relplot(x="ApplicantIncome", y="LoanAmount", hue = 'Gender', data = df)
```

### Plotting Categorical Data

```python
# we’ll see the relationship between two variables of which one would be categorical 
# Jitter Plot
sns.catplot(x="Gender", y="LoanAmount", hue = "Gender", data=df)
```

```python
sns.catplot(x="Gender", y="LoanAmount", kind = "box", data=df)
```

```python
# Histogram
sns.distplot(df.ApplicantIncome, kde=False, rug = True)
```
